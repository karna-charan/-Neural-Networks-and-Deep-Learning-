```python
# ---------------- INSTALL DEPENDENCIES ----------------
!pip install ultralytics opencv-python

# ---------------- IMPORTS ----------------
from ultralytics import YOLO
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow
import numpy as np
import os

# ---------------- UPLOAD IMAGE ----------------
uploaded = files.upload()
if not uploaded:
    raise RuntimeError("No image uploaded")

image_path = list(uploaded.keys())[0]  # get uploaded file name
print(f"Uploaded file: {image_path}")

# ---------------- LOAD IMAGE ----------------
image = cv2.imread(image_path)
if image is None:
    raise RuntimeError("Failed to load image")

# ---------------- LOAD YOLOv8 POSE MODEL ----------------
model = YOLO("yolov8n-pose.pt")  # will auto-download if needed

# ---------------- RUN POSE DETECTION ----------------
results = model(image)

# ---------------- BODY PART NAMES ----------------
BODY_PARTS = [
    "Nose",
    "Left Eye", "Right Eye",
    "Left Ear", "Right Ear",
    "Left Shoulder", "Right Shoulder",
    "Left Elbow", "Right Elbow",
    "Left Wrist", "Right Wrist",
    "Left Hip", "Right Hip",
    "Left Knee", "Right Knee",
    "Left Ankle", "Right Ankle"
]

# ---------------- ANALYZE & OUTPUT ----------------
found = False
print("\nDetected body parts:\n")

for r in results:
    if r.keypoints is None:
        continue

    keypoints = r.keypoints.xy[0]  # first person detected
    # keypoints is now a NumPy array of shape (17,2)
    found = True

    for i, point in enumerate(keypoints):
        x, y = point  # unpack correctly
        x, y = int(x), int(y)
        name = BODY_PARTS[i]
        print(f"{name:15} -> x:{x}, y:{y}")

        # draw on image
        cv2.circle(image, (x, y), 5, (0, 255, 0), -1)
        cv2.putText(image, name, (x, y - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 255, 0), 1)

if not found:
    print("No human body detected.")

# ---------------- SHOW RESULT ----------------
cv2_imshow(image)
```
