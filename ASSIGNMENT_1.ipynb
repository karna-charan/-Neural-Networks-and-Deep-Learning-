**7.Extract the second row and third column from a 2D tensor.**

```python
import torch

# Example 2D tensor
t = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

# Extract second row and third column
second_row = t[1]      # index 1 = second row
third_column = t[:, 2] # index 2 = third column

print("Second row:", second_row)
print("Third column:", third_column)
```

**8.Compute matrix multiplication between tensors of shape (3,4) and (4,2).**

```python
import torch

# Create tensors
A = torch.randn(3, 4)  # shape (3,4)
B = torch.randn(4, 2)  # shape (4,2)
# Matrix multiplication
C = torch.matmul(A, B)
print("Result:\n", C)
print("Shape:", C.shape)
```

**10.Implement softmax using tensor operations (no built-in function).**

```python
import torch
# Example tensor
x = torch.tensor([1.0, 2.0, 3.0])
# Softmax (manual)
x_max = torch.max(x)                 # for numerical stability
exp_x = torch.exp(x - x_max)         # exponentiate
softmax = exp_x / torch.sum(exp_x)   # normalize
print("Softmax:", softmax)
```

**11.Normalize a tensor so that values lie between 0 and 1.**

```python
import torch
# Example tensor
t = torch.tensor([2.0, 5.0, 10.0, 20.0])
# Min-Max normalization to [0, 1]
t_min = t.min()
t_max = t.max()
normalized = (t - t_min) / (t_max - t_min)
print("Normalized tensor:", normalized)
```

**12.Write code to compute gradients of a tensor operation (PyTorch)**

```python
import torch
# Create tensor with gradient tracking
x = torch.tensor(2.0, requires_grad=True)
# Tensor operation
y = x**2 + 3*x + 1   # y = x² + 3x + 1
# Compute gradients
y.backward()
# Print gradient dy/dx
print("Gradient:", x.grad)
```

**13.Convert an image tensor (H, W, C) to (C, H, W).**

```python
import torch
# Example image tensor with shape (H, W, C)
img = torch.randn(64, 64, 3)  # (H=64, W=64, C=3)
# Convert to (C, H, W)
img_chw = img.permute(2, 0, 1)
print("New shape:", img_chw.shape)
```

**14.Mask elements in a tensor based on a condition.**

```python
import torch
# Example tensor
t = torch.tensor([1, 5, 8, 2, 10])
# Mask elements greater than 5
mask = t > 5          # condition
masked = t[mask]      # apply mask
print("Mask:", mask)
print("Masked values:", masked)
```

**20.Implement cosine similarity using tensor operations.**

```python
import torch
# Example tensors
a = torch.tensor([1.0, 2.0, 3.0])
b = torch.tensor([4.0, 5.0, 6.0])
# Cosine similarity
dot = torch.dot(a, b)
norm_a = torch.norm(a)
norm_b = torch.norm(b)
cos_sim = dot / (norm_a * norm_b)
print("Cosine Similarity:", cos_sim)
```
