```python
import cv2
import numpy as np
from google.colab import files
from IPython.display import Image, display

# Upload image
uploaded = files.upload()

# Get uploaded file name
image_path = list(uploaded.keys())[0]

# Read image in color
img = cv2.imread(image_path)

# Check if image loaded
if img is None:
    print("Error: Image not found.")
else:
    # Take color choice from user
    color_choice = input("Which color to keep? (red/green/blue): ").lower()

    # Create blank image
    output = np.zeros_like(img)

    # OpenCV uses BGR format
    if color_choice == "red":
        output[:, :, 2] = img[:, :, 2]  # Red channel
    elif color_choice == "green":
        output[:, :, 1] = img[:, :, 1]  # Green channel
    elif color_choice == "blue":
        output[:, :, 0] = img[:, :, 0]  # Blue channel
    else:
        print("Invalid choice. Choose red, green, or blue.")
        exit()

    # Save the output
    output_filename = f"{color_choice}_channel_image.jpg"
    cv2.imwrite(output_filename, output)

    # Display the output
    display(Image(output_filename))
    print(f"{output_filename} saved successfully!")
```

```python
import cv2
import numpy as np
from google.colab import files
from IPython.display import Image, display

# Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Read image
img = cv2.imread(image_path)

if img is None:
    print("Error: Image not found.")
else:
    # Convert once for filters that need grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Define filters in a dictionary
    filters = {
        "blur": cv2.GaussianBlur(img, (5, 5), 0),
        "sobel_x": cv2.convertScaleAbs(cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)),
        "sobel_y": cv2.convertScaleAbs(cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)),
        "outline": cv2.Canny(gray, 100, 200),
        "grayscale": gray
    }

    # Show available filters
    print("Available filters:", ", ".join(filters.keys()))
    choice = input("Enter filter: ").lower()

    # Apply selected filter
    if choice in filters:
        output = filters[choice]
        output_filename = f"{choice}_image.jpg"
        cv2.imwrite(output_filename, output)
        display(Image(output_filename))
        print(f"{output_filename} saved successfully!")
    else:
        print("Invalid filter choice!")
```

```python
import cv2
from google.colab import files
from IPython.display import Image, display

# Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Read grayscale image
gray_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if gray_img is None:
    print("Error: Image not found.")
else:
    print("Choose coloring option:")
    print("1 - Normal BGR color")
    print("2 - Heatmap (JET colormap)")

    choice = input("Enter choice (1 or 2): ")

    # Apply selected option
    if choice == "1":
        color_img = cv2.cvtColor(gray_img, cv2.COLOR_GRAY2BGR)
    elif choice == "2":
        color_img = cv2.applyColorMap(gray_img, cv2.COLORMAP_JET)
    else:
        print("Invalid choice!")
        exit()

    # Save and display
    output_filename = "color_output.jpg"
    cv2.imwrite(output_filename, color_img)
    display(Image(output_filename))
    print("Image saved as:", output_filename)
```
